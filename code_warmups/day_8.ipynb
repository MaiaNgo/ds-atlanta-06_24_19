````python
def kmsq_2_milesq(dataframe, column_name, new_column_name):
    new_df = dataframe.copy()
    new_df[column_name] = new_df[column_name] * 0.621371 ** 2
    new_df.rename(columns={column_name: new_column_name}, inplace=True)
    
    return new_df

print(country_column)
kmsq_2_milesq(country_column, 'Area_km_sq', 'Area_mile_sq')
````
