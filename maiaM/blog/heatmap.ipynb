# What is Heatmap and How to use it in Python

### What is Heatmap?

###### A Heat Map is a graphical representations of data that utilize color-coded systems.
- The data values are contained in a matrix form
- Each data value is represented as a color in a color-coding scheme. The darker the shade, the    greater the quantity; the higher the value, the tighter the dispersion, or vice versa etc.
- Perfect for representing data where a defined boundary is relevant to the data (which is marked and bounded area on a map denoting the distribution of some property).
- Website tracking and tools to optimize conversions

### How to use Heatmap in Python using the Seaborn visiualization library? 

First, we need to import Seaborn library. And, remember Seaborn is built on top of Matplotlib, so we need to include the magic command matplotlib to show inline graphs in Jupyter notebook.

```python
# import Seaborn library
import seaborn as sns

# set this to show graph inline in Jupyter notebook
%matplotlib inline
```

Now, we load the 'flights' sample data set from the built-in Seaborn library sample data.

https://github.com/mwaskom/seaborn-data/blob/master/flights.csv

```python
# https://seaborn.pydata.org/generated/seaborn.load_dataset.html
# load sample 'flights' data set from the Seaborn online GitHub repositoty
flights = sns.load_dataset('flights')

# show the head of the DataFrame
flights.head()
```

The data table above shows the feature of number of passengers that flied in each month of each year. But the data table is not in matix form, we need to convert it into a matrix form with index is the months and columns is the years by using a Pivot table.

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot_table.html#pandas.DataFrame.pivot_table
# convert DataFrame to matrix format
flight_matrix = flights.pivot_table(index='month', columns='year', values='passengers')

# show the head of the DataFrame
flight_matrix
```

```python
#### TOTAL OF ALL CODES FOR EXAMPLE.

# import Seaborn library
import seaborn as sns

# set this to show graph inline in Jupyter notebook
%matplotlib inline

# https://seaborn.pydata.org/generated/seaborn.load_dataset.html
# load sample 'flights' data set from the Seaborn online GitHub repositoty
flights = sns.load_dataset('flights')

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot_table.html#pandas.DataFrame.pivot_table
# convert DataFrame to matrix format
flight_matrix = flights.pivot_table(index='month', columns='year', values='passengers')

# https://seaborn.pydata.org/generated/seaborn.heatmap.html
# plot heatmap and show data values
sns.heatmap(flight_matrix, cmap='coolwarm', annot=True, fmt="d")

# https://seaborn.pydata.org/generated/seaborn.heatmap.html
# plot heatmap
sns.heatmap(flight_matrix)

sns.heatmap(flight_matrix, cmap='YlGnBu')
```

Now we see the data in a matrix form where each data value (each cell) is the number of passengers flied for the month on that row and the year on that column.

With this matrix data, we can easily plot a heatmap for it using Seaborn's heatmap plot.

```python
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
# plot heatmap
sns.heatmap(flight_matrix)
```

In the Heatmap above, we see that the color-coding legend is showed on the right of the map. The colors go from dark to light which the higher the data values the lighter the colors.

But if we want, we can change to another color map to show from cool to warm, the higher the data values the wamrer the colors, then we can customize the heatmap as below.

```python
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
# plot heatmap with color map 'coolwarm' (defined in Matplotlib)
sns.heatmap(flight_matrix, cmap='coolwarm')
```

We can also choose whether or not we want to show the data value on each cell of the heatmap.

```python
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
# plot heatmap and show data values
sns.heatmap(flight_matrix, cmap='coolwarm', annot=True, fmt="d")
```

Now, we look at the heatmap of the dataset 'flights' again to see what it tells us.

```python
sns.heatmap(flight_matrix, cmap='YlGnBu')
```

- The very first thing we see is that the later the year the more people fly, this make sense because over time the flights is more accessible to people.
- The second thing we notice is that in any given year, the months of July and August tend to have much more people fly than other months. This also make sense because that the summer time where families tend to have more long vacation trips.
