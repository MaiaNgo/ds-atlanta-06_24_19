```python
data_dictionary = {"country": ["USA", "Canada", "Mexico", "Nicaragua", "Belize", "Guatamala", "El Salvador", "Costa Rica", "Panama", "Honduras"], "Population": [300000000, 150000000, 167000000, 87000000, 35000000, 76000000, 45000000, 67000000, 55000000, 38000000], "Area_km_sq": [10000000, 14000000, 4000000, 900000, 500000, 400000, 420000, 370000,390000, 420000]}
data_dictionary
```

```python
# Convert the dictionary into a DataFrame
import pandas as pd

# construct DataFrame from dictionary
countries_df = pd.DataFrame(data_dictionary)
countries_df
```

```python
# Set the country column as the dataframe index
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.set_index.html
countries_df.set_index('country')

# Notice that the change above is not affect the countries_df - because by default inplace=False
#print(countries_df)
```

```python
# Convert the Area_km_sq column to square miles (Change the column heading as well)
def km_sq_2_mile_sq():
    # 1 km = 0.621371 mile
    return (0.621371 ** 2)

# add new column Area_mile_sq
countries_df['Area_mile_sq'] = countries_df['Area_km_sq'] * km_sq_2_mile_sq()
print(countries_df)

# drop Area_km_sq column
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html#pandas.DataFrame.drop
countries_df.drop('Area_km_sq', axis=1)
```

```python
# Create a new column that contains the population per sq_mile
countries_df['Population_per_sq_mile'] = countries_df['Population'] / countries_df['Area_mile_sq']
print(countries_df)

# convert to int
countries_df['Population_per_sq_mile'] = countries_df['Population_per_sq_mile'].astype(int)
countries_df
```
